```python
from apiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow
```

```python
scopes = ['https://www.googleapis.com/auth/calendar']
```

```python
flow = InstalledAppFlow.from_client_secrets_file('client_secret.json',scopes=scopes)
credentials = flow.run_console()
```

```python
service = build('calendar', 'v3', credentials=credentials)
result = service.calendarList().list().execute()
```

```python
def create_events(title,typ,start,end,location=''):
    events = {
        'summary': f'({typ}) {title}',
        'description': location,
        'start': {
            'dateTime': start,
            'timeZone': 'Europe/Berlin',
        },
        'end': {
            'dateTime': end,
            'timeZone': 'Europe/Berlin',
        },
        'reminders': {
            'useDefault': False,
            'overrides': [
                {'method': 'popup', 'minutes': 10},
            ],
        },
    }
    return events
```

```python
import pandas as pd
from datetime import datetime
calendar_csv = pd.read_csv('personal_20230414_124709.csv')
calendar_csv.head()
```

```python
for index, row in calendar_csv.iterrows():
    title = row['TITEL']
    typ = row['LV_ART']
    location = row['ORT']
    date = row['DATUM'].split('.')
    start_time = row['VON'].split(':')
    start = datetime(int(date[2]),int(date[1]),int(date[0]),int(start_time[0]),int(start_time[1]))
    end_time = row['BIS'].split(':')
    end = datetime(int(date[2]),int(date[1]),int(date[0]),int(end_time[0]),int(end_time[1]))
    event = service.events().insert(calendarId=calendar_id, body=create_events(title,typ,start.isoformat(),end.isoformat(),location)).execute()
```
